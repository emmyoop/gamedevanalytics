### Compare results for two data cleanup methods to see if i miussed anything

```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv('consolidated_processed.csv')
df1 = pd.read_csv('initial_processed.csv')

df1 = df1.fillna('NULL')
df2 = df2.fillna('NULL')
```

```python
comparison = df1.eq(df2)
```

```python
matches = comparison.apply(pd.Series.value_counts)
matches
```

```python
# lets drop the columns that have 100% matching

non_match = matches.copy()
col_names = list(matches)

for index, row in matches.iterrows():
    for col in col_names:
        if np.isnan(row[col]):
            non_match.drop(columns=col, inplace=True)
    break

non_match
```

```python
# now let's pull out the columns that have 100% non-match
zero_match = non_match.copy()

# the only nan values left are those in the True row (ie, 0 matches between DFs)
col_names = list(non_match)
for index, row in non_match.iterrows():
    if index != 0:
        for col in col_names:
            if not(np.isnan(row[col])):
                zero_match.drop(columns=col, inplace=True)
zero_match

# these complete no matches are already acocunted for and can be ignored
```
